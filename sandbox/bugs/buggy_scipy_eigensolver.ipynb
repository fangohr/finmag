```python
N = 80
```

Define a matrix `A` for which we will compute eigenvalues and eigenvectors. This is a Laplacian-type matrix (with "periodic boundary conditions", since the upper right and lower left entry is set as well). The exact solutions are known (see [1], Lemma 2.4.4), and are simple sinusoidal functions.

[1] http://www.cs.yale.edu/homes/spielman/561/2009/lect02-09.pdf

```python
A = np.zeros((N, N))

for i in xrange(N):
    A[i, i] = 2.0
    A[i, i-1] = -1.0
    A[i, (i+1)%N] = -1.0

print A
```

We use a sparse Hermitian solver in Scipy to compute the lowest eigenvalues. To do this, we use the so-called "shift-and-invert" mode (specified by the arguments `sigma` and `which`).

The following code seems to work most of the time, but when it is executed ca. 15-20 times, every now and then the solutions will look very jittery and wrong (as in the example below). Interestingly, the computed solutions are also different each time the code is run, so there is some randomness hidden somewhere.

```python
# Solve the eigenproblem
omega, w_T = scipy.sparse.linalg.eigsh(A, sigma=0.0, which='LM')

# Sort the eigenvalues and eigenvectors
sort_indices = omega.argsort()
omega = omega[sort_indices]
w = w_T.T[sort_indices]

plot(w[4])
plot(w[3])
```
