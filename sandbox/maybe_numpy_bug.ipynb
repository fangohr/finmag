# A Bug in Numpy?

## Visualisation of What is Happening

Set up two arrays $x$ and $y$ with values from 0 to 1.

```python
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)
```

With

$z(x, y) = 1 + x * y$

we expect $z\in[0, 2].$

```python
Z = 1 + X * Y
print "Z has values from {} to {}.".format(np.min(Z), np.max(Z))
```

$z$ can be visualised.

```python
fig = figure(figsize=(4, 4))
ax = fig.add_subplot(111)
surf = ax.pcolormesh(X, Y, Z)
ax.set_title("z = 1 + x * y")
ax.set_xlabel("x")
ax.set_ylabel("y")
fig.colorbar(surf)
```

Now, let's substract $z(0, 0) = 1.$

```python
Z -= Z[0]
```

```python
fig = figure(figsize=(4, 4))
ax = fig.add_subplot(111)
surf = ax.pcolormesh(X, Y, Z)
ax.set_title("z = 1 + x + y")
ax.set_xlabel("x")
ax.set_ylabel("y")
fig.colorbar(surf)
```

The substraction isn't executed throughout the array!

## Minimal Failing Example

```python
a = linspace(0, 1, 100)
b = linspace(0, 1, 100)

A, B = meshgrid(a, b)
C = 1 + A * B
print "Values in C from {} to {}.".format(C.min(), C.max())
```

```python
print "Last row of C"
print C[-1]
```

```python
print "Substracting first row (all ones) of C from whole array."
C -= C[0]
```
