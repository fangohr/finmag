```python
m0 = [0, 4, -1.5]  # this direction seems to match pinned direction in paper figure

# However, it is easier to understand to align pinned and applied direction to 
# z and y direction.

# Choose bias direction in z.
m0 = [0, 0, 1]

print("bias direction is {:.2f} deg".format(math.atan2(m0[2], m0[1]) * 360. / 2 / math.pi))
```

Create simulation object

```python
Ms = 1150e3
from finmag.energies import UniaxialAnisotropy
sim = finmag.sim_with(mesh, m_init=m0, Ms=Ms, A=14e-12, unit_length=1e-9,
                      demag_solver=None)

# anisotropy to represent demag of infinite film in y-z
mu0 = 4 * math.pi * 1e-7
Kx = -0.5 * mu0 * Ms ** 2
sim.add(UniaxialAnisotropy(Kx, axis=[1, 0, 0], name='Kx'))
```

```python
def pin_fun(pos):
    x=pos[0]
    if x<1e-5:
        return True
    
sim.pins = pin_fun
```

```python
def plot_m(m, title, filename=None):
    """Given a dolfin magnetisation function m (or other function), 
    plot components mx, my, mz as a function of space. 
    Expect 1d-space (=mesh) in x-direction."""

    xs = mesh.coordinates()
    mx = [m(x)[0] for x in xs]
    my = [m(x)[1] for x in xs]
    mz = [m(x)[2] for x in xs]
    
    plt.plot(xs, mx, 'o-', label='mx')
    plt.plot(xs, my, 'x-', label='my')
    plt.plot(xs, mz, '^-', label='mz')
    plt.xlabel('x [nm]')
    plt.ylabel('normalised magnetisation components')
    plt.legend()
    plt.title(title)
    if filename:
        plt.savefig(filename)
```

```python
plot_m(sim.get_field_as_dolfin_function('m'), 'initial magnetisation','test.png')
```

```python
# additional code for 3d plots
def add_vector_to_axis(ax, vecstart, vecend, **kwds):    
    x, y, z = zip(vecstart, vecend)
    ax.plot(x, y, z, **kwds)
    
def render_m_on_1d_mesh(sim, ax, n = None):
    """n is the number of vectors to plot.
    sim the finmag simulation object
    ax the matplot lib axes to which the vectors should be added.
    """
    mesh = sim.mesh
    if n == None:
        xs = mesh.coordinates()
    else:
        assert isinstance(n, int)
        assert n > 0
        coordinates = mesh.coordinates()
        
        # normalise
        max_x = max(coordinates)
        xs = np.linspace(min(coordinates), max(coordinates), n)

    m = sim.llg._m
    
    # for every point in the mesh, plot a vector at the mesh node
    for x in xs:
        mx, my, mz = m(x)
        origin = [x/max_x, 0, 0]
        if x <= 0:
            color = 'k'  # black if in bias-pinned layer
        else:
            color = 'b'  # blue otherwise
        add_vector_to_axis(ax, origin, [mx + x/max_x, my, mz], color=color)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')

    ax.view_init(elev=30, azim=5)
    
    ax.set_zlim3d(0, 1)                    # viewrange for z-axis should be [-1, 1] 
    ax.set_ylim3d(-1, 0.3)                    # viewrange for y-axis should be [-1, 

    return ax
```

```python
import numpy
m_app_dir = numpy.array([0, -m0[2], m0[1]])  # perpendicular to m0
# normalise:
m_app_dir /= numpy.linalg.norm(m_app_dir) 
print("field direction is {}, that is {:.2f} deg".format(m_app_dir,
            math.atan2(m_app_dir[2], m_app_dir[1]) * 360. / 2 / math.pi))
assert numpy.linalg.norm(m_app_dir) == 1.0
```

```python
def angles_between_vectors(reference, list_of_vectors, scale = 360. / 2. /math.pi):
    ref = reference
    results = []
    for vec in list_of_vectors:
        
        # angle to pinned_m0 and app_field
        cos_alpha = numpy.dot(ref, vec) / (norm(ref) * norm(vec))
        alpha = math.acos(cos_alpha)
        results.append(alpha * scale)
    return results
```

# Simulate the experiment

```python
# range of applied fields in milli Tesla
Bs = numpy.array([0, 1, 3, 4, 8, 16])

print("Range of fields in mT: {}".format(Bs))
```

```python
# convert Bs into A/m
Hs = Bs * 1e-3 / mu0
```

```python
for H, B in zip(Hs, Bs):
    print("Working on B = {}mT".format(B))
    filename='m-{:04}mT'.format(B)
    if os.path.exists(filename + '.npy'):
        continue
    sim.set_H_ext(m_app_dir * H)
    finmag.set_logging_level('INFO')
    sim.relax()
    
    # magnetisation at top (i.e. where imaging takes place in experiment)
    mtop = sim.get_field_as_dolfin_function('m')(max(sim.mesh.coordinates()))
    alpha0, alpha1 = angles_between_vectors(mtop, [m0, m_app_dir])
    print("Angle to pinned layer:  {:.2f} deg".format(alpha0))
    print("Angle to applied field: {:.2f} deg".format(alpha1))
    #assert 90-1e5 < alpha0 + alpha1 < 90 +1e-5   # sanity check

    plt.figure() # create new figure
    title = 'B = {} mT, angle to pinned = {:.2f} deg'.format(B, alpha0)
    # 2d plot
    plot_m(sim.get_field_as_dolfin_function('m'), title)
           #filename='{}.png'.format(B))

    # 3d plot
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax = render_m_on_1d_mesh(sim, ax, n=30)
    #plt.show()

    # save data
    sim.save_m(filename=filename, overwrite=True)
    
```

## Create the plot from saved data

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#fig = plt.figure(figsize=(12,16))
fig = plt.figure(figsize=(7,10))

for H, B, i, letter in zip(Hs, Bs, range(1, 7), ['(a)', '(b)', '(c)', '(d)', '(e)', '(f)']):
    print("Working on B = {}mT".format(B))
    filename='m-{:04}mT.npy'.format(B)
    m = np.load(filename)
    sim.set_m(m)
    
    ax = fig.add_subplot(320 + i, projection='3d')
   
    # magnetisation at top (i.e. where imaging takes place in experiment)
    mtop = sim.get_field_as_dolfin_function('m')(max(sim.mesh.coordinates()))
    alpha0, alpha1 = angles_between_vectors(mtop, [m0, m_app_dir])
    print("Angle to pinned layer:  {:.2f} deg".format(alpha0))
    print("Angle to applied field: {:.2f} deg".format(alpha1))
    #assert 90-1e5 < alpha0 + alpha1 < 90 +1e-5   # sanity check

    subfigureletter = '{}'.format(letter)
    tmp = '$\mathbf{' + subfigureletter + '} \quad\quad'
    title = tmp + r' B = {}\,{},\quad \phi_\mathrm{{top}} = {:.1f}^\circ$'.format(B, '\mathrm{mT}', alpha0)
    ax.set_title(title)
    ax.set_xlabel('$z / d$')
    render_m_on_1d_mesh(sim, ax, n=30)
    ax.set_xlabel('$z$', fontsize=14)
    ax.set_zlabel(r'<$\vec{m} | \vec{u}_\mathrm{exchange-bias}>$', fontsize=14)
    ax.set_ylabel(r'<$\vec{m} | \vec{u}_\mathrm{applied}>$', fontsize=14)
    
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['0', 'd'])
   
    ax.set_yticks([-1, 0]) #actually -1, 0
    ax.set_yticklabels(['1', '0']) #actually -1, 0, but looks better if its 1, 0
    ax.set_zticks([0, 1])
    ax.view_init(elev=30, azim=20)
    # stop after first plot for development
    #break 
```
