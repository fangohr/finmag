An ipython notebook to report some strange 'blips' which occur in the total energy (the meaning of 'blip' will become apparent when you see the plots below!). I'm kind of thinking they're related to the problem which only occurs on the core-i7's.

The simulation is a 40x40nm 2D mesh, with exchnage, DMI and applied field.

A single skyrmion is initialised in the centre and the simulation is then run for 2ns.

```python
import finmag
from finmag import sim_with
import numpy as np
import dolfin as df
from finmag import Simulation
from finmag.energies import Exchange, Demag, DMI, Zeeman
from finmag.sim.magnetisation_patterns import initialise_skyrmions
from finmag.util.helpers import set_logging_level, fnormalise

import pylab

set_logging_level("INFO")
```

```python
# set up simulation objects

x0 = y0 = -20
x1 = y1 = 20
xn = yn = 40

mesh2D = df.RectangleMesh(x0, y0, x1, y1, xn, yn,"crossed")
    
Ms = 0.157e6 
A = 3.53e-13 	
D = 0.00025
H = Zeeman((0, 0, 282600))
unit_length = 1e-9
alpha = 0.05
    
sim_name = 'sim__energy_blips_investigation'
sim = Simulation(mesh2D, Ms=Ms, unit_length=unit_length, name=sim_name,pbc="2d")
sim.set_tol(reltol=10e-11,abstol=10e-11)
    
initialise_skyrmions(sim,skyrmionRadius=10)
    
sim.add(Exchange(A))
sim.add(DMI(D))
sim.add(H)
sim.alpha = alpha

sim.schedule('save_ndt', every=1e-12)
sim.run_until(2.0e-9) 
```

```python
data = finmag.util.fileio.Tablereader('sim__energy_blips_investigation.ndt')

t = data['time']
E_t = data['E_total']
E_dmi = data['E_DMI']
E_ex = data['E_Exchange']
E_zee = data['E_Zeeman']
dmdt_x = data['dmdt_x']
dmdt_y = data['dmdt_y']
dmdt_z = data['dmdt_z']
dmdt = np.sqrt(dmdt_x**2 + dmdt_y**2 + dmdt_z**2)
```

Plotting the total energy, it decays as expected (woo!)

```python
pylab.plot(t[0:2000],E_t[0:2000],'.',label = 'E total - 2D system')
fig = pylab.gcf()
fig.set_size_inches(12,8)
pylab.legend()
```

Zooming in on a section now shows these strange blips in the energy.

```python
pylab.plot(t[1000:2000],E_t[1000:2000],'.',label = 'E total - 2D system')
fig = pylab.gcf()
fig.set_size_inches(12,8)
pylab.legend()
```

```python
pylab.plot(t,dmdt,'.', label = 'dmdt 2D')
pylab.legend()
```
