```python
import math
import numpy as np
import matplotlib as mpl
mpl.rcParams['font.size'] = 12
import matplotlib.pyplot as plt

pi = math.pi
```

# Discrete Fourier Transforms in 1D with Numpy

```python
def transform(n, xn):
    k = np.fft.fftfreq(len(n), n[1] - n[0])
    Xk = np.fft.fft(xn)
    
    # Returned in standard order, which
    # matplotlib sorts automatically when plotting.
    # Use np.fft.fftshift to sort by
    # increasing frequency.
    return k, Xk
    
def show_fourier(n, xn, normalise=False, show_all=True):
    k, Xk = transform(n, xn)
    amplitude = np.abs(Xk)
    if normalise:
        amplitude = 2 * amplitude / len(n)
    power = np.abs(Xk) ** 2
    phase = np.angle(Xk)
        
    fig = plt.figure(figsize=(16, 4))
    nb_subplots = 4 if show_all else 2
    
    ax = fig.add_subplot(1, nb_subplots, 1)
    ax.set_title("Realspace Data")
    ax.set_xlabel("time")
    ax.set_ylabel("signal")
    ax.plot(n, xn, color="blue", linewidth=0.2)
    ax.plot(n, xn, "bo", markersize=4, label="$x_n$")
    ax.legend()
     
    ax = fig.add_subplot(1, nb_subplots, 2)
    ax.set_title("Amplitude Spectrum")
    ax.set_xlabel("frequency")
    ax.set_ylabel("amplitude")
    ax.plot(k, amplitude, color="blue", linewidth=0.2)
    ax.plot(k, amplitude, "bo", markersize=4, label="$|X_k|$")
    ax.legend()
    
    if show_all:
        ax = fig.add_subplot(1, nb_subplots, 3)
        ax.set_title("Power Spectrum")
        ax.set_xlabel("frequency")
        ax.set_ylabel("power")
        ax.plot(k, power, color="blue", linewidth=0.2)
        ax.plot(k, power, "bo", markersize=4, label="$|X_k|^2$")
        ax.legend()
        
        ax = fig.add_subplot(1, nb_subplots, 4)
        ax.set_title("Phase Spectrum")
        ax.set_xlabel("frequency")
        ax.set_ylabel("coefficients")
        ax.plot(k, phase, color="blue", linewidth=0.2)
        ax.plot(k, phase, "bo", markersize=4, label="angle $X_k$")
        ax.legend()
    
    fig.tight_layout()
    return fig
    
```

```python
n = np.linspace(0, 5, 256)
xn = np.sin(2 * pi * n)
_ = show_fourier(n, xn)
```

To original amplitude is obtained by multiplication of the discrete Fourier amplitude with $\frac{2}{N}$.

```python
k, Xk = transform(n, xn)
amplitude = np.abs(Xk)
amplitude_normalised = 2 * amplitude / len(n)
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(111)
ax.set_title("Amplitude Spectrum")
ax.set_xlabel("frequency")
ax.set_ylabel("amplitude (normalised)")
ax.plot(k, amplitude_normalised, color="blue", linewidth=0.2)
ax.plot(k, amplitude_normalised, "bo", markersize=4, label="$|X_k|$")
ax.legend()
```

## Parseval's Theorem

$$\sum_{n=0}^{N-1}\left|x_n\right|^2 = \frac{1}{N} \sum_{k=0}^{N-1}\left|X_k\right|^2$$

```python
n  = np.linspace(0, 5, 257)
xn = np.sin(2 * pi * n)
parseval_left = np.sum(np.abs(xn) ** 2)
print parseval_left
```

```python
k, Xk = transform(n, xn)
scaling_term = 1.0 / len(n)
parseval_right = scaling_term * np.sum(np.abs(Xk) ** 2)
print parseval_right
```

```python
difference = parseval_left - parseval_right
print difference
```

## Examples

### Gaussian Function

```python
def bell_curve(n, a=1.0, b=0.0, c=1.0):
    """
    Return bell curve over points `n`.
    Params: height `a`, pos. of centre `b`, FWHM `c`.

    """
    return a * np.exp(- (n - b)**2 / (2 * c**2))

n = np.linspace(-50, 50, 100)
```

By varying the width of the bell curve, we can see the uncertainty principle in action. The wider the function is, the more concentrated the spectrum becomes, and vice-versa.

```python
fig = show_fourier(n, bell_curve(n, c=2.0), show_all=False)
fig.axes[0].set_ylim((0, 2))
fig.axes[1].set_ylim((0, 50))
```

```python
fig = show_fourier(n, bell_curve(n, c=20.0), show_all=False)
fig.axes[0].set_ylim((0, 2))
fig.axes[1].set_ylim((0, 50))
```

### Square Signal

We will progressively add sine waves to approximate a square function and see the new freqencies appear in the spectrum.

```python
n = np.linspace(-2, 2, 100)
xn = np.zeros(len(n))
for k in xrange(5):
    xn += np.sin((2 * k + 1) * pi * n) / (2 * k + 1)
    _ = show_fourier(n, xn, normalise=True, show_all=False)
```

### One-Sided, Real, Decaying Exponential Signal

```python
n = np.linspace(-3, 3, 100)
s = np.exp(-n)
s[np.where(n<0)] = 0
```

```python
_ = show_fourier(n, s)
```
