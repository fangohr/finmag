# Spin wave propagation in periodic system

```python
import finmag
import dolfin as df
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri
from finmag import Simulation as Sim
from finmag.energies import Exchange, Demag
finmag.set_logging_level('INFO')
%matplotlib inline
```

The mesh in this example is a three-dimensional stripe with edge lengths $a$ and $b$ and thickness $d$. The spacing of vertices in this regular mesh is 1 nm along all edges.

```python
a = 90
b = 30
d = 1
lmax = 1

mesh = df.BoxMesh(0, 0, 0, a, b, d, a/lmax, b/lmax, d/lmax)
```

The used material is Permalloy with the following parameters (saturation magnetsiation $M_\text{s}$, exchange constant $A$, and Gilbert damping $\alpha$):
$$M_\text{s} = 10^6 \text{A/m}$$
$$A = 13 \times 10^{-12} \text{J/m}$$
$$\alpha = 0.02$$

```python
Ms = 1e6  # magnetisation saturation (A/m)
A = 13e-12  # exchange stiffness (J/m)
alpha = 0.02  # Gilbert damping
```

Now, the simulation object is created and exchange and demagnetisation energies are added to the simulation. In addition, one-dimensional periodic boundary conditions are posed.

```python
sim  = Sim(mesh, Ms, unit_length=1e-9, pbc='1d')  # create simulation object

# Add energies.
sim.add(Exchange(A))
sim.add(Demag())

# Set the Gilbert damping value in the LLG equation.
sim.alpha = alpha
```

The initial magnetisation is not defined at this point. Since the spin waves should occur, system has to be initialised in an appropriate way. In this case, the magnetisation is uniform at all mesh nodes, except in a circular region at the left edge of the boundary. The function which will be used for magnetisation initialisation is:

```python
def m_initial(coord):
    # Extract x and y coordinates.
    x = coord[0]
    y = coord[1]
    
    # The centre of the circle
    x_centre = a/10.
    y_centre = b/2.
    
    # Compute the circle radius.
    r = ((x-x_centre)**2 + (y-y_centre)**2)**0.5
    
    if r < 2:
        return (1, 0, 0.1)
    else:
        return (1, 0, 0)
    
sim.set_m(m_initial)
```

The function used for plotting magnetisation $z$ component in all subsequent plots.

```python
def plot_magnetisation(sim):
    # Extract the mesh node coordinates.
    mesh = sim.mesh
    coords = mesh.coordinates()
    
    # Sample the magnetisation at the bottom of the mesh.
    mz = []
    coords_bottom = []
    for coord in coords:
        if coord[2] == 0:
            coords_bottom.append(coord)
            mz.append(sim.llg._m(coord)[2])
            
    x = [noname[0] for noname in coords_bottom]
    y = [noname[1] for noname in coords_bottom]
    triang = tri.Triangulation(x, y)
    
    # Plot.
    plt.figure(figsize=(10, 4))
    plt.tripcolor(triang, mz, shading='gouraud')
    plt.xlabel('x (nm)')
    plt.ylabel('y (nm)')
    plt.grid()
    
plot_magnetisation(sim)
```

When the initial magnetisation is set, the simulation is executed for 0.1 ns, and the magnetisation is plotted every $\Delta t = 50 \text{ps}$

```python
t_sim = 2e-12  # simulation time (s)
dt = 0.25e-12  # plotting time step (s)
t_array = np.arange(0, t_sim, dt)

sim.reset_time(0)
for t in t_array:
    sim.run_until(t)
    plot_magnetisation(sim)
```

It can be seen that the spin wave "travels across" the left boundary in $x$ direction.
