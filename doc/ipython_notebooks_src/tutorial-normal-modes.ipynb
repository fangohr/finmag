# Computation and visualisation of normal modes

## Generating some sample data

In this example we compute and visualise the normal mode of a magnetic rod. We use the well-established method of letting the magnetisation precess around an external field which is applied along the long axis of the rod. During the precession we save the magnetisation at regular time intervals and compute the Fourier transform to determine the normal modes.

### Stage 1: Relaxation

In the first stage we apply the external field along an axis that is slightly off the main axis of the rod and let the system relax.

```python
import finmag
from finmag.example import barmini
from finmag.energies import Zeeman
from finmag.util.visualization import render_paraview_scene
from finmag.util.meshes import box
from finmag import sim_with
finmag.set_logging_level('INFO')
```

```python
mesh = box(0, 0, 0, 100, 100, 5, maxh=10)
print mesh.num_vertices()
```

```python
#sim = sim_with(mesh, Ms=8e5, m_init=(0, 0, 1), alpha=1.0
```

```python
sim = barmini()
H = Zeeman([0, 0.001e6, 1e6])
sim.add(H)
```

Let's do quick visualisation of the mesh by saving a `.vtu` file and rendering it using Paraview:

```python
sim.save_vtk(filename='snapshots/mymesh.pvd', overwrite=True)
render_paraview_scene('snapshots/mymesh000000.vtu', 'snapshots/mymesh.png', camera_position=[150, 150, -80], camera_view_up=[1, 0, 0],
                      add_glyphs=False, show_colorbar=False, show_orientation_axes=True, view_size=(600, 400))
```

```python
sim.relax()
```

```python
sim.save_restart_data('barmini_relaxed.npz')
```

```python
sim.save_vtk('snapshots/relaxed_state.pvd', overwrite=True)
render_paraview_scene('snapshots/relaxed_state000000.vtu', 'snapshots/relaxed_state.png', camera_position=[150, 150, -80], camera_view_up=[1, 0, 0],
                      rescale_colormap_to_data_range=False, add_glyphs=True, show_colorbar=True, show_orientation_axes=True, view_size=(600, 400))
```

### Phase 2: Precession

Just so that we have a clean slate, let's restart the simulation and reset the time to zero:

```python
sim.restart('barmini_relaxed.npz', t0=0)
sim.clear_schedule()
```

We re-set the external field so that it now points exactly along the z-axis. Then we schedule saving of the field data at regular intervals (every 10 ps) and let the simulation run. (**Note:** This may take a while!)

```python
t_step = 2e-11
sim.alpha = 0.001
sim.set_H_ext([0, 0, 1e6])
sim.schedule('save_field', 'm', filename='m_values/m.npy', overwrite=True, every=t_step)
sim.schedule('save_ndt', every=t_step)
#sim.schedule('save_vtk', filename='snapshots/precession.pvd', every=1e-11)
sim.run_until(5e-9)
```

## Visualizing the normal modes

```python
from finmag.util.fft import *
```

### Plotting the frequency spectrum

Finmag has a convenience function to compute the Fourier transforms of the (spatially averaged) magnetisation components. This can be useful for post-processing or plotting. We illustrate how to use it to plot the spectra. 

```python
fft_freqs, fft_mx, fft_my, fft_mz = FFT_m('barmini.ndt', t_step=1e-11, subtract_values='average')
```

```python
plot(fft_freqs, fft_mx, '-')
plot(fft_freqs, fft_my, '-')
```

There is a convenience function which does the two steps automatically (and provides a bit of syntactic sugar to tweak the resulting plot):

```python
plot_FFT_m('barmini.ndt', t_step=1e-11, subtract_values='average', components='xy', xlim=(0, 20), figsize=(20, 4))
```

Let's extract the index `k` of the peak in the Fourier spectrum of $m_x$.

```python
k = fft_mx.argmax()
f_peak = fft_freqs[k]
print "Index of peak frequency: {}".format(k)
print "Peak frequency: {:.2f} GHz".format(f_peak / 1e9)
```

### Plotting a spatially resolved normal mode

```python
from glob import glob
from finmag.util.helpers import probe
```

**Note:** Using the functions explained in this section is still very clumsy and the interface is in heavy need of cleanup. Some of the issues are due to the fact that we are currently saving data in separate `.npy` files and cannot easily associate timestep information with each stored data vector. These should be mostly resolved once we have support for saving data to `.h5` files. Also, some of these functions might be better wrapped up in higher-level convenience functions. Moreover, probing the magnetisation is rather slow (at least for large meshes) due to current limitations in dolfin. Finally, All in all, the animation method in the next section is by far preferable.

First we need to probe all the saved magnetization vectors on a regular grid.

```python
# All this code is far too low-level and should be wrapped up.

def npy_to_dolfin_function(V, filename):
    a = np.load(filename)
    fun = df.Function(V)
    fun.vector().set_local(a)
    return fun

def probing_grid(mesh, nx, ny, nz):
    coords = mesh.coordinates()
    xmin, xmax = min(coords[:, 0]), max(coords[:, 0])
    ymin, ymax = min(coords[:, 1]), max(coords[:, 1])
    zmin, zmax = min(coords[:, 2]), max(coords[:, 2])
    grid = np.mgrid[xmin:xmax:nx*1j, ymin:ymax:ny*1j, zmin:zmax:nz*1j]
    return grid
```

```python
files = sorted(glob('./m_values/m_*.npy'))  # note to self: 'sorted' is super-important!
print "Number of files found: {}".format(len(files))
```

```python
dolfin_funcs = [npy_to_dolfin_function(sim.llg.S3, f) for f in files]
X, Y, Z = probing_grid(sim.mesh, 1, 10, 10)
```

```python
pts = empty(X.shape + (3,))
pts[..., 0] = X
pts[..., 1] = Y
pts[..., 2] = Z
probed_vals = np.ma.masked_array([probe(f, pts) for f in dolfin_funcs])
```

After all this clumsy preparation, we can finally plot the spatially resolved normal mode. Note that due to the small size and coarse resolution of the bar this isn't very meaningful, but at least it provides pretty pictures. ;-)

```python
_ = plot_spatially_resolved_normal_modes(probed_vals, k, t_step=t_step, yshift_title=2.5, figsize=(16, 5), show_colorbars=False)
```
