In Finmag, two NEB classes (NEB and NEB_Sundials) are implemented to provide the nudged elastic band method. Using some extra library such as pele (http://pele-python.github.io/pele/) is going to be included in Finmag as well. It is suggested to try NEB_Sundials first.

In this example, we will illustrate how to use NEB_Sundials to find the minimum energy path for a one-dimensional wire. To simplify the model only the uniaxial anisotropy and exchange interactions are considered. We know that it will favor the path with domain wall rather than the coherent rotation path if the wire is long enough.

As usual, let's start by importing Finmag.

```python
import numpy as np
import dolfin as df
from finmag import Simulation as Sim
from finmag.energies import Exchange, UniaxialAnisotropy
from finmag.physics.neb import NEB_Sundials
from finmag.util.helpers import set_logging_level
set_logging_level("INFO")
```

We create the simulation as normal.

```python
def create_simulation(mesh):
    
    sim = Sim(mesh, Ms=8.6e5, unit_length=1e-9)
    
    A = 1.3e-11
    Kx = 8e4
    
    sim.set_m((1,0,0))
    
    sim.add(Exchange(A))
    sim.add(UniaxialAnisotropy(Kx, (1, 0, 0), name='Kx'))
    
    return sim
```

Suppose the initial magnetization distribution is (-1,0,0) and final distribution is (1,0,0), we give some hints to NEB_Sundials by providing a DW-like intermediate state, which is given by the following function.

```python
def dw_1d(pos):
    x = pos[0]
    
    if x<400:
        return (1, 0, 0)
    elif x>600:
        return (-1, 0, 0)
    else:
        return (0, 1, 0)
```

Then we can start to use the NEB_Sundials,.

```python
def relax_neb(sim):
    
    init_images=[(-1, 0, 0), dw_1d, (1, 0, 0)]
    # the initial images number between each initial image
    interpolations = [10, 10]
    
    neb = NEB_Sundials(sim, init_images, interpolations, spring=1e6, name='neb')

    neb.relax(max_steps=1000, save_vtk_steps=200, stopping_dmdt=1e2)
```

The most important parameters in `neb.relax()` method are `max_steps` and `stopping_dmdt`, let's run it.

```python
mesh = df.IntervalMesh(500, 0, 1000)
```

```python
sim = create_simulation(mesh)
```

```python
relax_neb(sim)
```
